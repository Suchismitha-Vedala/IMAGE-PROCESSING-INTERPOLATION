```python
I=data
```

```python
a=0
C=np.zeros(shape=(x,y))
for i in range(x):
    for j in range(y): 
        C[i][j]=a 
        a=a+1
```

```python
x,y=I.shape
R=np.zeros(shape=(x,y))
k=1
i=1
j=1
for i in range(x):
    for j in range(y):
        if (I[i][j]==1 and I[i][j-1]==0):
            if (I[i-1][j]==0):
                R[i][j]=k
                k=k+1
            elif (I[i-1][j]==1):
                R[i][j]=R[i-1][j]
        elif (I[i][j]==1 and I[i][j-1]==1):
            if (I[i-1][j]==0):
                R[i][j]=R[i][j-1]
            elif (I[i-1][j]==1):
                R[i][j]=R[i-1][j]
        if (R[i][j]==R[i][j]/R[i-1][j]):
            R[i][j]=R[i-1][j]
            
```
