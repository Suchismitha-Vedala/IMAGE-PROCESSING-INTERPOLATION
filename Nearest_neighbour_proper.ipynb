```python
img = cv2.imread("/Users/suchi/Desktop/Fall 2017/DIP/GIt/cells.png")
plt.imshow(img)
plt.show()
```

```python
w,h,c = img.shape
```

```python
fx=2
fy=2
```

```python
new_w=int(w*fx)
new_h=int(h*fy)
```

```python
b=np.zeros((new_h,new_w,c),np.uint8)
```

```python
for i in range(new_w):
    for j in range(new_h):
        b[j,i,:]=img[int(i/fx)][int(j/fy)][:]
         
```
